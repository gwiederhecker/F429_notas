```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
freq=np.logspace(0,7,20)
R= 1e3
rl = 50
L=47e-3
Clist = np.array([0.01,1,22,47,470,1000,100000])*1e-9
omega=2*np.pi*freq
for C in Clist:
    Zc=-1j/(omega*C)
    Zl=1j*omega*L+rl
    Zt=R+Zc+Zl
    H=R/(Zt)
    T=np.abs(H)**2
    fig=plt.plot(freq,10*np.log10(T+2e-3/1),'-')
    plt.grid()
    plt.xscale('log')
    plt.ylim([-40,0])
# freq=np.logspace(1,7,1000)
# omega=2*np.pi*freq
# for C in Clist:
#     Zc=-1j/(omega*C)
#     Zl=1j*omega*L+rl
#     Zt=R+Zc+Zl
#     H=R/(Zt)
#     T=np.abs(H)**2
#     plt.plot(freq,10*np.log10(T),'k-')
plt.show()
```
